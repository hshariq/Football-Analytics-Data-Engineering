```python
response = requests.get('https://api.sofascore.com/api/v1/event/11369489/statistics', headers=headers)
```

```python
stats = response.json()
```

```python
len(stats['statistics'])
```

```python
#create a stats1 such that if period is all it removes it, else keeps it in stas1
stats1 = [x for x in stats['statistics'] if x['period'] != 'a']
```

```python
stats1
```

```python
len(stats1)
```

```python
for x in stats1:
    print(x['period'])
    if x['period'] == '1ST':
        for y in x['groups']:
            y['period'] = '1st'
    elif x['period'] == '2ND':
        for y in x['groups']:
            y['period'] = '2nd'
    elif x['period'] == 'ALL':
        for y in x['groups']:
            y['period'] = 'ALL'
```

```python
#merge both periods into one period 
stats1 = stats1[0]['groups'] + stats1[1]['groups'] + stats1[2]['groups']
```

```python
stats1
```

```python
for x in stats1:
    print(x['period'])
```

```python
import pandas as pd
# Create a dictionary to store DataFrames for each group
dataframes_by_group = {}

# Iterate over the list and create DataFrames for each group
for stats_group in stats1:
    group_name = stats_group['groupName']
    statistics_items = stats_group['statisticsItems']
    period = stats_group['period']
    
    # Create a DataFrame for each group
    df = pd.DataFrame(statistics_items)
    df['period'] = period

    if group_name in dataframes_by_group:
        # If the group already exists, append the new DataFrame to the current DataFrame
        dataframes_by_group[group_name] = pd.concat([dataframes_by_group[group_name], df])
    else:
        # If the group doesn't exist, create a new DataFrame with the current DataFrame
        dataframes_by_group[group_name] = df
    
# Access DataFrames by group name
for group_name, df in dataframes_by_group.items():
    print(f"Group Name: {group_name}")
    print(df)
    print("\n")
```
